## Editing Atoms with RDKit

In addition to seeing information about atoms and bonds in our molecule, we can also use RDKit to change our molecule structure.
Let's consider benzene for our example. We will change one of the carbons in the benzene ring to a nitrogen to make pyridine.

We are going to create a copy of our benzene molecule using another function from RDKit called `RWMol`. `RWMol` makes our molecule readable and writeable (or "editable").

````python
# TO DO: we will create pyridine from benzene, an RDKit mol object we already created in this noteobook

pyridine = Chem.RWMol(benzene)

pyridine                            # but you can see it's still benzene, not pyridine
````

One way we can change molecules in RDKit is to set the atomic number of a particular atom to another number.
To change one of the carbons in our ring to nitrogen, we can select one of the atoms and then set its atomic number to that of nitrogen (7).

````python
# TO DO: Get atom with index 0 (labeled 0 above)

atom = pyridine.GetAtomWithIdx(0)

# TO DO: Set the atomic number to 7

atom.SetAtomicNum(7)

pyridine
````

## Combining two molecules

We can also create and combine molecules using RDKit. This approach might be necessary when we want to add a more complex functional group. In the example below we create a carboxyl group and add it to benzene to make benzoic acid.

First we create the molecule for our functional group.

````python
carboxyl = Chem.MolFromSmiles('C(=O)O')
carboxyl
````

Next, we combine the two molecules together using the `CombineMols` function.

````python
combined = Chem.CombineMols(benzene, carboxyl)
combined
````

After the combining, we have to make our combined molecules editable so that we can change the bonds and atoms.

````python
editable_mol = Chem.RWMol(combined)
editable_mol
````

Finally, we can use `AddBonds` to add a bond between the carbon on our carboxyl group (6) to an atom on benzene. We could have picked any carbon, but the example below uses index 5.

````python
editable_mol.AddBond(0, 6, order=Chem.rdchem.BondType.SINGLE)
````

````python
editable_mol
````

# Three Powerful Approaches to Molecular Visualization with RDKit + py3Dmol

## Approach 1: Single Molecule Deep Dive
**When to use**: Detailed exploration of ONE molecule's structure and properties

### Key Features:
- **Multiple visual layers**: Combines stick model, ball representation, and semi-transparent surfaces
- **Automatic annotation**: Labels heteroatoms (N, O, S, etc.) for quick identification
- **Dynamic presentation**: Spinning animation for 360° viewing
- **Property profiling**: Calculates molecular descriptors (MW, LogP, H-bonding)

### Best for:
- Teaching/presentations
- Detailed structural analysis
- Understanding specific molecular features
- Publication-quality figures

**Example**: Caffeine visualization with VDW surface and Jmol coloring

---

## Approach 2: Comparative Multi-Molecule Gallery
**When to use**: Comparing MULTIPLE DIFFERENT molecules side-by-side

### Key Features:
- **Grid layout**: 2×2 or 3×2 panels for simultaneous comparison
- **Unique styling per molecule**: Different color schemes distinguish each structure
- **Synchronized viewing**: All panels rotate together for easier comparison
- **Drug-likeness screening**: Lipinski's Rule of Five analysis for each molecule
- **Comprehensive property table**: MW, LogP, TPSA, rotatable bonds, H-bonding

### Best for:
- SAR (Structure-Activity Relationship) studies
- Lead compound selection
- Scaffold comparison
- Building molecular libraries

**Example**: Aspirin, Ibuprofen, Morphine, and Acetaminophen comparison with drug-likeness analysis

---

## Approach 3: Conformational Ensemble Analysis
**When to use**: Exploring different 3D shapes of the SAME molecule

### Key Features:
- **Conformer generation**: Creates 10+ different 3D arrangements (rotamers)
- **Energy-based ranking**: Sorts conformers by thermodynamic stability
- **Force field optimization**: Each conformer is energy-minimized
- **Boltzmann statistics**: Shows population distribution at room temperature
- **Color-coded stability**: Blue (stable) → Green (medium) → Orange (unstable)
- **Energy landscape mapping**: Visualizes conformational space

### Best for:
- Understanding molecular flexibility
- Protein-ligand docking preparation
- Pharmacophore modeling
- QSAR studies requiring 3D descriptors
- Predicting bioactive conformations

**Example**: Tamiflu conformers with energy analysis showing the molecule can adopt multiple shapes, with the most stable conformer populated ~90% at 298K

---

## Quick Decision Guide

| Your Goal | Use Approach |
|-----------|--------------|
| "What does this drug look like?" | **Approach 1** |
| "How do these compounds differ?" | **Approach 2** |
| "What shapes can this molecule adopt?" | **Approach 3** |
| "Is this drug-like?" | **Approach 2** |
| "Which conformer binds to the protein?" | **Approach 3** |
| "Explain this structure to students" | **Approach 1** |

---

## Pro Tips

1. **Approach 1** is perfect for posters and slides - it's visually striking and informative
2. **Approach 2** scales well - you can easily expand to 3×3 or 4×4 grids
3. **Approach 3** is essential for flexible molecules with rotatable bonds (> 5 rotatable bonds)

### Combining Approaches:
You can mix them! For example:
- Use **Approach 3** to find the lowest-energy conformer
- Then use **Approach 1** to showcase that conformer in detail
- Finally use **Approach 2** to compare it with related compounds

**Remember**: The right visualization depends on your scientific question! 🔬✨

````python
from rdkit.Chem import AllChem, Descriptors
from IPython.display import display
````

````python
# Create an interesting molecule - let's use Caffeine, a familiar stimulant
caffeine_smiles = "CN1C=NC2=C1C(=O)N(C(=O)N2C)C"
mol = Chem.MolFromSmiles(caffeine_smiles)

# Add hydrogens and generate 3D coordinates
mol = Chem.AddHs(mol)
AllChem.EmbedMolecule(mol, randomSeed=42)
AllChem.MMFFOptimizeMolecule(mol)

# Convert to Mol block for visualization
mol_block = Chem.MolToMolBlock(mol)

# Create a stunning 3D visualization with multiple representations
viewer = py3Dmol.view(width=800, height=600)

# Add the molecule with a stick representation
viewer.addModel(mol_block, 'mol')

# Style 1: Ball and stick with beautiful colors
viewer.setStyle({'stick': {'radius': 0.15, 'color': 'spectrum'},
                 'sphere': {'scale': 0.3, 'colorscheme': 'Jmol'}})

# Add semi-transparent surface to show molecular volume
viewer.addSurface(py3Dmol.VDW, {'opacity': 0.7, 'colorscheme': 'whiteCarbon'})

# Add labels for heteroatoms
for atom in mol.GetAtoms():
    if atom.GetSymbol() not in ['C', 'H']:
        pos = mol.GetConformer().GetAtomPosition(atom.GetIdx())
        viewer.addLabel(atom.GetSymbol(),
                       {'position': {'x': pos.x, 'y': pos.y, 'z': pos.z},
                        'backgroundColor': 'black',
                        'fontColor': 'white',
                        'fontSize': 12})

# Set a nice viewing angle and zoom
viewer.zoomTo()
viewer.rotate(45, {'x': 1, 'y': 1, 'z': 0})

# Enable spin for dramatic effect
viewer.spin(True)

display(viewer)

# Print some molecular properties for context
print(f"Molecule: Caffeine (C₈H₁₀N₄O₂)")
print(f"Molecular Weight: {Descriptors.MolWt(mol):.2f} g/mol")
print(f"LogP: {Descriptors.MolLogP(mol):.2f}")
print(f"Number of H-bond donors: {Descriptors.NumHDonors(mol)}")
print(f"Number of H-bond acceptors: {Descriptors.NumHAcceptors(mol)}")
````

````python
from rdkit.Chem import Lipinski
import numpy as np
````

````python
# Create a pharmacologically interesting set: Common painkillers
molecules = {
    "Aspirin": "CC(=O)Oc1ccccc1C(=O)O",
    "Ibuprofen": "CC(C)Cc1ccc(cc1)C(C)C(=O)O",
    "Morphine": "CN1CC[C@]23[C@@H]4[C@H]1CC5=C2C(=C(C=C5)O)O[C@H]3[C@H](C=C4)O",
    "Acetaminophen": "CC(=O)Nc1ccc(cc1)O"
}

# Create a multi-panel viewer
viewer = py3Dmol.view(width=1000, height=700, viewergrid=(2,2))

# Color schemes for different molecules
color_schemes = ['greenCarbon', 'cyanCarbon', 'magentaCarbon', 'orangeCarbon']

for idx, (name, smiles) in enumerate(molecules.items()):
    # Generate 3D structure
    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol, randomSeed=42)
    AllChem.MMFFOptimizeMolecule(mol, maxIters=200)

    # Remove hydrogens for cleaner visualization
    mol = Chem.RemoveHs(mol)
    mol_block = Chem.MolToMolBlock(mol)

    # Calculate row and column for grid
    row, col = idx // 2, idx % 2

    # Add to viewer at specific grid position
    viewer.addModel(mol_block, 'mol', viewer=(row, col))

    # Apply unique styling to each molecule
    viewer.setStyle({'stick': {'radius': 0.2, 'colorscheme': color_schemes[idx]},
                     'sphere': {'scale': 0.25, 'colorscheme': 'Jmol'}},
                    viewer=(row, col))

    # Add molecular orbital-like surface
    viewer.addSurface(py3Dmol.SAS,
                     {'opacity': 0.6, 'colorscheme': color_schemes[idx]},
                     viewer=(row, col))

    # Add title label
    viewer.addLabel(name,
                   {'position': {'x': 0, 'y': 0, 'z': 5},
                    'backgroundColor': 'darkblue',
                    'fontColor': 'white',
                    'fontSize': 14,
                    'fontOpacity': 0.9},
                   viewer=(row, col))

    # Calculate and print properties
    mol_with_h = Chem.AddHs(mol)
    print(f"\n{'='*50}")
    print(f"🔬 {name}")
    print(f"{'='*50}")
    print(f"Formula: {Chem.rdMolDescriptors.CalcMolFormula(mol_with_h)}")
    print(f"MW: {Descriptors.MolWt(mol):.2f} g/mol")
    print(f"LogP: {Descriptors.MolLogP(mol):.2f}")
    print(f"TPSA: {Descriptors.TPSA(mol):.2f} Ų")
    print(f"Rotatable Bonds: {Descriptors.NumRotatableBonds(mol)}")
    print(f"H-bond Donors: {Descriptors.NumHDonors(mol)}")
    print(f"H-bond Acceptors: {Descriptors.NumHAcceptors(mol)}")

    # Lipinski's Rule of Five
    violations = Lipinski.NumHDonors(mol) > 5 or \
                 Lipinski.NumHAcceptors(mol) > 10 or \
                 Descriptors.MolWt(mol) > 500 or \
                 Descriptors.MolLogP(mol) > 5
    print(f"Lipinski Violations: {'❌ Failed' if violations else '✅ Passed'}")

    viewer.zoomTo(viewer=(row, col))

# Sync rotation across all panels for dramatic effect
viewer.rotate(30, {'x': 1, 'y': 1, 'z': 0})

display(viewer)

print(f"\n{'='*25}")
print("All molecules displayed in interactive 3D grid!")
print("Each has unique coloring and surface representation")
````

````python
# Let's use Oseltamivir (Tamiflu) - an antiviral with interesting conformational flexibility
tamiflu_smiles = "CCOC(=O)C1=CC(CC(C1)NC(=O)C)N(CC)CC"
mol = Chem.MolFromSmiles(tamiflu_smiles)
mol = Chem.AddHs(mol)

# Generate multiple conformers to explore conformational space
print("Generating conformational ensemble...")
conformer_ids = AllChem.EmbedMultipleConfs(mol, numConfs=10,
                                            randomSeed=42,
                                            pruneRmsThresh=0.5,
                                            useRandomCoords=True)

# Optimize each conformer and get energies
print("Optimizing conformers and calculating energies...")
energies = []
for conf_id in conformer_ids:
    props = AllChem.MMFFGetMoleculeProperties(mol)
    ff = AllChem.MMFFGetMoleculeForceField(mol, props, confId=conf_id)
    ff.Minimize()
    energy = ff.CalcEnergy()
    energies.append(energy)

# Sort by energy
sorted_indices = np.argsort(energies)
relative_energies = np.array(energies) - min(energies)

# Create a spectacular multi-conformer visualization
viewer = py3Dmol.view(width=1200, height=800, viewergrid=(2,3))

# Show top 6 conformers
for idx in range(min(6, len(conformer_ids))):
    conf_id = int(sorted_indices[idx])  # ← FIX: Convert to Python int
    row, col = idx // 3, idx % 3

    # Get this conformer
    conf_mol = Chem.Mol(mol)
    conf_mol.RemoveAllConformers()
    conf_mol.AddConformer(mol.GetConformer(conf_id))
    conf_mol = Chem.RemoveHs(conf_mol)
    mol_block = Chem.MolToMolBlock(conf_mol)

    # Add to viewer
    viewer.addModel(mol_block, 'mol', viewer=(row, col))

    # Color by energy: low energy = blue, high energy = red
    energy_ratio = relative_energies[conf_id] / max(relative_energies) if max(relative_energies) > 0 else 0

    if energy_ratio < 0.3:
        color_scheme = 'cyanCarbon'
        emoji = '❄️'  # Cold = stable
    elif energy_ratio < 0.7:
        color_scheme = 'greenCarbon'
        emoji = '🌿'  # Medium
    else:
        color_scheme = 'orangeCarbon'
        emoji = '🔥'  # Hot = unstable

    # Style based on stability
    viewer.setStyle({'stick': {'radius': 0.15, 'colorscheme': color_scheme},
                     'sphere': {'scale': 0.3, 'colorscheme': 'Jmol'}},
                    viewer=(row, col))

    # Add electrostatic-like surface
    viewer.addSurface(py3Dmol.VDW,
                     {'opacity': 0.5, 'colorscheme': color_scheme},
                     viewer=(row, col))

    # Label with conformer info
    label_text = f"Conformer {idx+1} {emoji}\n+{relative_energies[conf_id]:.1f} kcal/mol"
    viewer.addLabel(label_text,
                   {'position': {'x': 0, 'y': 0, 'z': 4},
                    'backgroundColor': 'rgba(0,0,0,0.7)',
                    'fontColor': 'white',
                    'fontSize': 11},
                   viewer=(row, col))

    viewer.zoomTo(viewer=(row, col))

# Add rotation for drama
viewer.rotate(45, {'x': 1, 'y': 1, 'z': 0})

display(viewer)
````

````python
# Print detailed analysis
print(f"\n{'='*40}")
print(f"CONFORMATIONAL ANALYSIS: Oseltamivir (Tamiflu)")
print(f"{'='*40}\n")

print(f" Statistical Summary:")
print(f"   • Total conformers generated: {len(conformer_ids)}")
print(f"   • Energy range: {max(relative_energies):.2f} kcal/mol")
print(f"   • Most stable conformer: #{sorted_indices[0]+1}")
print(f"   • Absolute minimum energy: {min(energies):.2f} kcal/mol\n")

# Boltzmann distribution at room temperature
RT = 0.592  # kcal/mol at 298K
boltzmann_weights = np.exp(-relative_energies / RT)
boltzmann_weights /= boltzmann_weights.sum()

print(f"  Boltzmann Population (298K):")
for idx in range(min(6, len(conformer_ids))):
    conf_id = sorted_indices[idx]
    pop_percent = boltzmann_weights[conf_id] * 100
    bar = '=' * int(pop_percent / 2)
    print(f"   Conformer {idx+1}: {pop_percent:5.1f}% {bar}")

print(f"\n Molecular Properties:")
mol_no_h = Chem.RemoveHs(mol)
print(f"   • Formula: {rdMolDescriptors.CalcMolFormula(mol)}")
print(f"   • Molecular Weight: {Descriptors.MolWt(mol_no_h):.2f} g/mol")
print(f"   • LogP: {Descriptors.MolLogP(mol_no_h):.2f}")
print(f"   • TPSA: {Descriptors.TPSA(mol_no_h):.2f} Ų")
print(f"   • Rotatable Bonds: {Descriptors.NumRotatableBonds(mol_no_h)}")
print(f"   • Chiral Centers: {len(Chem.FindMolChiralCenters(mol_no_h, includeUnassigned=True))}")

# Ring analysis
print(f"\n💍 Ring System Analysis:")
ring_info = mol_no_h.GetRingInfo()
print(f"   • Number of rings: {ring_info.NumRings()}")
print(f"   • Ring sizes: {[len(ring) for ring in ring_info.AtomRings()]}")

print(f"\n✨ The colors represent conformational energy:")
print(f"   ❄️  CYAN = Most stable (lowest energy)")
print(f"   🌿 GREEN = Intermediate stability")
print(f"   🔥 ORANGE = Least stable (highest energy)")
print(f"\n{'='*80}")
print("Each conformer shows a different possible 3D shape of the same molecule!")
print("The Boltzmann distribution tells you which shapes are most populated at room temp.")
````
